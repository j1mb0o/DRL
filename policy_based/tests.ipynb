```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

list = [y1, y2, y3]
# Create a figure and axis object
fig, ax = plt.subplots()

# Loop over the data and create a plot for each dataset
for y in list:
    ax.plot(x, y)

# Set axis labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Multiple plots')

# Show the plot
plt.show()
```

```python
import os 
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
# import scienceplots
# plt.style.use('science')


dir = 'run_results'
# print(os.listdir(dir))

results = [np.load(os.path.join(dir, file)) for file in os.listdir(dir)]
result_name = [file.split('.npy')[0] for file in os.listdir(dir)]
print(result_name)
x_linspace = np.linspace(1, 2000, 2000)
print(x_linspace)

for (result, result_name) in zip(results, result_name):
    learning_rate = result_name.split('_')[2]
    gamma = result_name.split('_')[3]

    result = savgol_filter(result, 101, 3)
    plt.plot(x_linspace, result.mean(axis=0), label=f'={gamma}, ={learning_rate}')
    plt.fill_between(x_linspace, result.mean(axis=0) - result.std(axis=0), result.mean(axis=0) + result.std(axis=0), alpha=0.3)
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Actor Critic Experiments')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
```

```python
import matplotlib.pyplot as plt
import numpy as np 
import scienceplots

plt.style.use('science')

N = 21
x = np.linspace(0, 10, 11)
y = [3.9, 4.4, 10.8, 10.3, 11.2, 13.1, 14.1,  9.9, 13.9, 15.1, 12.5]

# fit a linear curve and estimate its y-values and their error.
a, b = np.polyfit(x, y, deg=1)
y_est = a * x + b
y_err = x.std() * np.sqrt(1/len(x) +
                          (x - x.mean())**2 / np.sum((x - x.mean())**2))

fig, ax = plt.subplots()
ax.plot(x, y_est, '-')
ax.fill_between(x, y_est - y_err, y_est + y_err, alpha=0.2)
ax.plot(x, y, 'o', color='tab:brown')

plt.figure(figsize=(20,20))
```
